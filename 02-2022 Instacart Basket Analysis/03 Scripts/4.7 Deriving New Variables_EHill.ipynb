# 03 User-Defined Variables

### If/else statment

```python
#define price_label variable
def price_label(row):
        if row['prices'] <= 5:
            return 'Low-range product'
        elif (row['prices'] > 5) and (row['prices']<= 15):
            return 'Mid-range product'
        elif row['prices'] > 15:
            return 'High-range product'
        else: return 'not enough data'
```

```python
#drop the merge and pandas index columns
df = df.drop(columns=['Unnamed: 0', '_merge'])
```

```python
# apply price_label variable by creating a column named price_range
# axis=1 makes this user defined variable apply to rows
df['price_range'] = df.apply(price_label, axis=1)
```

### If-statment w/ loc() function

```python
# using loc() function to define where a variable is =<> integer print 'this'
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# same as above, next parameter
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
# same as above, with last parameter
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# count of each value in the variable
df['price_range_loc'].value_counts(dropna = False)
```

```python
#Q2 redo busiest day column so that 2 busiest days, 2 slowest days, else are returns
#create empty list 'result'
result_1 = []

#for loop, keep in mind colon is 'then', == is comparison operator
for value in df_op_merged['orders_day_of_week']:
    if value ==0 or value==1:
        result_1.append('busiest day')
    elif value ==4 or value==3:
        result_1.append('least busy')
    else:
        result_1.append('regularly busy')
```

```python
# create new column called 'busiest_days' and use result_1 list from for-loop
df_op_merged['busiest_days']=result_1
```
