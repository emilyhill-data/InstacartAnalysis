### Q4 Create loyalty flag using the transform() and loc() functions.

```python
#create max_order column finding max of user_order_number grouped by user_id
df_opm['max_order']=df_opm.groupby(['user_id'])['user_order_number'].transform(np.max)
```

```python
df_opm.head()
```

```python
#create loyalty flag using loc() on max_value column
#>40
df_opm.loc[df_opm['max_order']>40,'loyalty_flag']='loyal customer'
```

```python
#loyalty flag <=40 & >10
df_opm.loc[(df_opm['max_order']<=40) & (df_opm['max_order']>10),'loyalty_flag']='regular customer'
```

```python
#loyalty flag <=10
df_opm.loc[df_opm['max_order']<=10, 'loyalty_flag']='new customer'
```

```python
#check loyalty flag for frequencies
df_opm['loyalty_flag'].value_counts(dropna=False)
```

### Q6 create a spending flag for each user based on the avg price across all orders using:
#### grouping by user_id, mean prices of products < 10, then 'low spender'
#### grouping by user_id, mean prices of products >=10, then 'high spender'

```python
# create average product price column
df_opm['average_product_price']=df_opm.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#spending flag<10
df_opm.loc[df_opm['average_product_price']<10,'spending_flag']='low spender'
```

```python
#spending flag >=10
df_opm.loc[df_opm['average_product_price']>=10,'spending_flag']='high spender'
```

```python
# check spending flag for frequencies
df_opm['spending_flag'].value_counts(dropna = False)
```

### Q7 create an order frequency flag for each user based on median 'days_since_prior_order' across all orders using:
#### grouping by user_id, median 'days_since_prior_order'>20, then 'non-freq customer'
#### grouping by user_id, median 'days_since_prior_order'<=20 & >10, then 'reg customer'
#### grouping by user_id, median 'days_since_prior_order'<=10, then 'freq customer'

```python
#create median days since prior order column
df_opm['median_dspo']=df_opm.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#order freq flag >20
df_opm.loc[df_opm['median_dspo']>20, 'order_freq_flag']='non-freq cust'
```

```python
#order freq flag <=20 & >10
df_opm.loc[(df_opm['median_dspo']<=20) & (df_opm['median_dspo']>10), 'order_freq_flag']='reg cust'
```

```python
#order freq flag <=10
df_opm.loc[df_opm['median_dspo']<=10, 'order_freq_flag']='freq cust'
```

```python
#check order freq flag for frequencies
df_opm['order_freq_flag'].value_counts(dropna = False)
```
