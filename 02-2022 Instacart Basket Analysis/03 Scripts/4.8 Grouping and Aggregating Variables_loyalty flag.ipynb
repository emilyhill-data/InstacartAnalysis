# 03 Grouping Data Lesson

### only grouping/no action

```python
df.groupby('product_name')
```

### grouping w single aggregation 

```python
#this is the average user order number grouped by dept id
#this is the general format for both single and multiple aggregations
#add other aggregate computations by keeping them inside the brackets (to index those aggregations)
df.groupby('department_id').agg({'user_order_number': ['mean']})
```

```python
#quick notation to aggregate a single value, not as pretty
df.groupby('department_id')['user_order_number'].mean()
```

```python
#to understand what average/mean is being computed 
df.describe()
```

### grouping w multiple aggregations

```python
# performing multiple aggregations, use this format:
df.groupby('department_id').agg({'user_order_number': ['mean', 'min', 'max', 'count']})
```

```python
#counts of customer orders grouped by user id sorted by user_id
df.groupby('user_id').agg({'user_order_number':['count']})
```

```python
#counts of customer orders grouped by user id, sorted by count desc
df['user_id'].value_counts(dropna=False)
```

```python
#create new column w groupby and returning a max-value
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['user_order_number'].transform(np.max)
```
